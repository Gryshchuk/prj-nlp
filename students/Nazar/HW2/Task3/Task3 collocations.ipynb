```python
verbs=["say", "tell", "speak", "claim", "communicate"]
```

```python
generatew_verbs = [dictionary.synonym(word) for word in  verbs]
generatew_verbs
```

```python
import numpy as np
all_verbs = list(set(verbs+ list(np.concatenate(generatew_verbs))))
all_verbs
```

```python
import spacy
from tqdm import tqdm
nlp = spacy.load('en')
```

```python
def contains_verbs(text, dictionary):
    entity_types=[]
    for item in nlp(text):
        if ((item.pos_=='VERB') and (item.lemma_ in dictionary)):
            return True
    return False


sentenses_with_verbs=[]
for i in tqdm(range(0,len(sents))):
    sent=sents[i]
     
    if ('ly' in sent) and contains_verbs(sent, all_verbs):
        sentenses_with_verbs.append(sent)
```

```python
sentenses_with_verbs_and_adv_ly=[]
result={}
for i in tqdm(range(0,len(sentenses_with_verbs))):
    sent= sentenses_with_verbs[i]
    for token in nlp(sent):
        if ((token.dep_=='advmod') and (token.text.endswith('ly')) and(token.head.lemma_ in all_verbs)):
            sentenses_with_verbs_and_adv_ly.append(sent)
            #print(token.text, token.head.lemma_)
            if token.head.lemma_ not in result:
                result[token.head.lemma_]=[]
            result[token.head.lemma_].append(token.text.lower())
            
```

```python
import collections
```

```python
for verb in result:
    collections.Counter(result[verb])
    print("{0}:{1}".format(verb,collections.Counter(result[verb]).most_common(10)))
    print('------------------------------------')
```
